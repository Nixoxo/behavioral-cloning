```python
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
import numpy as np

from six.moves import cPickle as pickle
import numpy as np
import pandas as pd
from scipy import stats, integrate

import preprocess
import model
```

```python
origin_pickle = "origin.pickle"
delta_pickle = "delta.pickle"

def load_data():
    pickle_1 = open(origin_pickle, 'rb')
    data_1 = pickle.load(pickle_1)
    pickle_2 = open(delta_pickle, 'rb')
    data_2 =pickle.load(pickle_2)

    X1 = data_1['features']
    Y1 = data_1['labels']
    X2 = data_2['features']
    Y2 = data_2['labels']

    X = np.append(X1, X2, axis=0)
    Y = np.append(Y1, Y2, axis=0)
    
    return X, Y    
```

```python
features, labels = load_data()
```

```python
ax = plt.axes()
x = pd.Series(labels, name="Steering angle")
sns.distplot(x, bins=51, kde=False, norm_hist=False,ax = ax)

ax.set_title('Distribution of Steering Angle')
ax.set_ylabel('Count')
plt.show()
```
